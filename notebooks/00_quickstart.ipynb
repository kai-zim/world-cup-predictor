# WC2026 Predictor — Quickstart

Synthetic end-to-end run: Elo replay -> goal model -> Monte-Carlo bracket simulation.

```python
import sys; sys.path.insert(0, '../src')
import numpy as np, pandas as pd
from wc2026.config.schema import AppConfig
from wc2026.pipeline import run_forecast

rng = np.random.default_rng(0)
teams = [f'Team{i:02d}' for i in range(32)]
dates = pd.date_range('2022-12-19', periods=500, freq='3D')
rows = []
for d in dates:
    a, b = rng.choice(teams, 2, replace=False)
    sa, sb = int(a[4:]), int(b[4:])
    rows.append(dict(date=d, tournament='Friendly', home_team=a, away_team=b,
        home_score=int(rng.poisson(max(0.3, 2.0-sa*0.05))),
        away_score=int(rng.poisson(max(0.3, 2.0-sb*0.05))), neutral=True))
hist = pd.DataFrame(rows)
cfg = AppConfig(); cfg.simulation.n_simulations = 10000
forecast, ratings, bracket = run_forecast(hist, teams, cfg)
forecast.head(12)
```
